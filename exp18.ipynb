```python
# Importing the required libraries...
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Code 1: Random Seed and Fixing A Value
import numpy as np
import pandas as pd

np.random.seed(0)
df = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [0, 70, 30, 90, 60],
    'Sales': np.random.randint(100, 500, 5),
    'Profit': np.random.randint(10, 100, 5)
})

print(df.head())
```

```python
# Code 2: Area Plot
plt.figure()
plt.fill_between(df['Category'], df['Values'], color='green', alpha=0.25)
plt.title("Nikunj")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show()
```

```python
# Code 3: Sales and Profit Area Plot
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid") # here we are applying the seaborn style.
plt.figure(figsize=(7,4))
plt.fill_between(df['Category'], df['Sales'],
                 color='skyblue', alpha=0.5, label='Sales')

plt.fill_between(df['Category'], df['Profit'],
                 color='orange', alpha=0.5, label='Profit')

plt.title("Nikunj")
plt.xlabel("Category")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
# Code 4: Pie Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.2f%%')
plt.title("Nikunj")
plt.show()
```

```python
# Code 5: Donut Chart
plt.figure()
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%')
centre_circle = plt.Circle((0,0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Nikunj")
plt.show()
```

```python
# Code 6: Boxplot Of Sales (Outliers Detection)
plt.figure()
sns.boxplot(x=df['Values'])
plt.title("Nikunj")
plt.show()
```

```python
# Code 7: Perfect Correlation Heatmap
plt.figure()
corr = df[['Sales', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Nikunj")
plt.show()
```

```python
# Code 8: Correlation Heatmap with no preset condition
import seaborn as sns
import matplotlib.pyplot as plt

# Here we are only going to select the numeric columns.
corr = df[['Values','Sales','Profit']].corr()

plt.figure(figsize=(5,4))

sns.heatmap(corr, annot=True)

plt.title("Nikunj")

plt.show()
```

```python
# Code 9: Bubble Plot (3 Variable Chart)
# bubble plot is nothing but a scatter plot only.

import matplotlib.pyplot as plt

plt.scatter(df['Sales'], df['Profit'],
            s=df['Values']*10)
# here the variable s is fixed to denote the size of bubbles in the bubble plot.
plt.title("Nikunj")
plt.xlabel("Sales")
plt.ylabel("Profit")

plt.show()
```

```python
# Code 10: Bubble Plot Using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.scatterplot(x='Sales', y='Profit',
                size='Values',        # bubble size
                hue='Values',         # color
                data=df,
                sizes=(50, 300),      # min-max size
                palette='viridis')

plt.title("Nikunj")

plt.show()
```
